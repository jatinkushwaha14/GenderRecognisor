```python
from tensorflow import keras
import cv2 as cv
import mediapipe as mp
import time as time
model = keras.models.load_model('gender_recogniser.h5')
```

```python
mp_face_detection = mp.solutions.face_detection
mp_drawing = mp.solutions.drawing_utils
face_detection=mp_face_detection.FaceDetection(model_selection=1, min_detection_confidence=0.5)
```

```python
cap=cv.VideoCapture(0)
t0=0
while True:
    success,img=cap.read()
    img=cv.flip(img,1)
    imgrgb=cv.cvtColor(img,cv.COLOR_BGR2RGB)
    results=face_detection.process(imgrgb)
    if results.detections:
      for detection in results.detections:
        data=detection.location_data.relative_bounding_box
        h,w,c=img.shape
        x1,y1,x2,y2=int(data.xmin*w),int(data.ymin*h),int((data.xmin+data.width)*w),int((data.ymin+data.height)*h)
        roi=img[y1:y2,x1:x2]
        if roi.size == 0:
          continue
        small = cv.resize(roi, (128, 128))
        small = small.astype('float32') / 255.0
        small=small.reshape(1,128,128,3)
        ypred=model.predict(small)
        if(ypred>0.5):
            cv.rectangle(img,(x1-5,y1-50),(x1+105,y1-5), (0,255,0),-1)
            cv.putText(img, "WOMAN", (x1, y1-20), cv.FONT_HERSHEY_PLAIN, 2,(0,0,0),3)
        else:
            cv.rectangle(img,(x1-5,y1-50),(x1+70,y1-5), (0,0,255),-1)
            cv.putText(img, "MAN", (x1, y1-20), cv.FONT_HERSHEY_PLAIN, 2,(255,255,255),3)
    t1=time.time()
    fps=1/(t1-t0)
    cv.putText(img, str(int(1/(t1 - t0)))+ "fps", (20, 700), cv.FONT_HERSHEY_PLAIN, 2,(255,255,255),3)
    t0=t1
    cv.imshow("Output",img)
    if cv.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
```
